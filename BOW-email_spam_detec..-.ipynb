```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\kaaagle_Data\spam2.csv")
df
```

```python
df = df.replace({'ham' : 0 , 'spam' : 1})
```

```python
df.Category.value_counts()
df.Category.nunique()
```

```python
x = df.Message
y = df.Category
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x = v.fit_transform(x)
```

```python
v.vocabulary_
```

```python
dir(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_predict,y_test))
```

```python
emails = [
    "Congratulations! You have won a $1000 Amazon gift card. Click here to claim your prize now!",
    "Please find the attached project report and let me know your feedback by Friday."
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('nb' , MultinomialNB())
])
clf.fit(x_train,y_train)
```

```python
clf_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(clf_pred,y_test))
```
